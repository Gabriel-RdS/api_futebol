```python
def create_tabela_campeonato(tabela_response):
    """Função que cria um dataFrame de alguma tabela de campenato de futebol.
        1) Essa função recebe como parâmetro a resposta de uma requisição de uma tabela da Api do Futebol
    """

    
    tabela_br_json = json.loads(tabela_response.text)
    dados = []
    for row in range(len(tabela_br_json)):
        dados.append(
        [tabela_br_json[row]['posicao'],
        tabela_br_json[row]['time']['nome_popular'],
        tabela_br_json[row]['jogos'],
        tabela_br_json[row]['vitorias'],
        tabela_br_json[row]['empates'],
        tabela_br_json[row]['derrotas'],
        tabela_br_json[row]['gols_pro'],
        tabela_br_json[row]['gols_contra'],
        tabela_br_json[row]['saldo_gols'],
        tabela_br_json[row]['aproveitamento'],
        tabela_br_json[row]['variacao_posicao'],
        tabela_br_json[row]['ultimos_jogos'][0:5]
        ])
    
    tabela_br_json = pd.DataFrame(dados).rename(columns={0:"Posição", 1:"Time", 2:"Pontuação", 
                                                         3:"Nº Jogos", 4:"V", 5:"E", 6:"D", 7:"Gols Pro", 
                                                         8:"Gols Contra", 9:"Aproveitamento %", 
                                                         10:"Variação de Posição", 11:"Últimos Jogos"})
    return tabela_br_json
```
